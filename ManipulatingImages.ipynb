```python
# Loading images from a directory
root_dir = "../Image"
images = []
for imgs in os.listdir(root_dir):
    im = image.imread(root_dir+'/'+imgs)
    images.append(im)
    print(im.shape,imgs)
```

```python
# Saving the image
s_img = Image.open('../Image/Testimage.jpg')
# save to PNG format
s_img.save('../Image/Testimage.png', format='PNG')
saved_image = Image.open('../Image/Testimage.png')
print(saved_image.format)
```

```python
#Converting image to grayscale
g_img = s_img.convert(mode='L')
g_img = g_img.save('../Image/g_img.jpg',format='JPEG')
new_g_img = Image.open('../Image/g_img.jpg')
new_g_img
```

```python
#Resizing Images
saved_image.thumbnail((100,100))
print(saved_image.size)
saved_image
```

```python
# Using resize() function
resized_image = s_img.resize((600,600))
print(resized_image.size)
resized_image
```

```python
# Flipping Image

h_flip = resized_image.transpose(Image.FLIP_LEFT_RIGHT)
v_flip = resized_image.transpose(Image.FLIP_TOP_BOTTOM)
pyplot.subplot(131)
pyplot.imshow(resized_image)
pyplot.subplot(132)
pyplot.imshow(h_flip)
pyplot.subplot(133)
pyplot.imshow(v_flip)
```

```python
pyplot.subplot(311)
pyplot.imshow(resized_image.rotate(0))
pyplot.subplot(312)
pyplot.imshow(resized_image.rotate(45))
pyplot.subplot(313)
pyplot.imshow(saved_image.rotate(90))
```

```python
# Cropping the image
#The crop function takes a tuple argument that defines the two x/y coordinates of the box to crop out of the image. For example, if the image is 2,000 by 2,000 pixels,
# we can clip out a 100 by 100 box in the middle of the image by defining a tuple with the top-left and bottom-right points of (950, 950, 1050, 1050).

# creating a cropped image
cropped = saved_image.crop((100, 100, 200, 200))
# show cropped image
cropped
```
